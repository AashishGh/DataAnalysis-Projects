### Importing pandas, loading csv file and visualizing it.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
mydata=pd.read_csv("BanknoteAuthentication.csv")
mydata
```

```python
v1=mydata["V1"]
v2=mydata["V2"]
v1_v2=np.column_stack((v1,v2))
mean=np.mean(v1_v2,0)
std_dev=np.std(v1_v2,0)
print(mean)
print(std_dev)
plt.scatter(v1,v2,s=3)
plt.scatter(mean[0],mean[1],s=200,alpha=0.5)
plt.text(mean[0],mean[1],"Mean Value")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
kmresult=KMeans(n_clusters=2).fit(v1_v2)
clusters=kmresult.cluster_centers_
plt.scatter(v1,v2,s=3)

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
kmresult=KMeans(n_clusters=2).fit(v1_v2)
clusters=kmresult.cluster_centers_

plt.figure(figsize=(10,10))
plt.scatter(v1,v2, c= kmresult.labels_.astype(float), s=20, alpha=0.5)
plt.scatter(clusters[:,0],clusters[:,1],s=1000,alpha=0.5)
plt.xlabel("V1")
plt.ylabel("V2")
print(clusters)
[plt.text(clusters[i][0],clusters[i][1],"k"+str(i+1)) for i in range(2)]

plt.savefig("Resultgraph.png")
```
