```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import numpy as np
xys=np.array([[10,20],[30,45],[45,66],[12,11]])
mean=np.mean(xys,0)
std_dev=np.std(xys,0)


ellipse=patches.Ellipse([mean[0],mean[1]],2*std_dev[0],2*std_dev[1],alpha=0.25)

fig,graph =plt.subplots()
graph.scatter(xys[:,0],xys[:,1])
graph.scatter(mean[0],mean[1])
graph.add_patch(ellipse)
dists=[np.linalg.norm(xy-mean) for xy in xys]
print(dists)
min=np.min(xys,0)
max=np.max(xys,0)
normed=(xys-min)/(max-min)
print(xys)
print(normed)

xs=[xy[0] for xy in xys]
ys=[xy[1] for xy in xys]
for i in range(len(xys)):
    x=[mean[0],xs[i]]
    y=[mean[1],ys[i]]
    plt.plot(x,y)
    
```
